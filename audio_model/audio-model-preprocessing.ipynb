```python
import os
from tqdm import tqdm
```

## Get Filenames

```python
utterances_dir = "../mmsd_raw_data/utterances_final"
(_, _, filenames) = next(os.walk(utterances_dir), (None, None, []))
```

## Convert Files to .wav format (you do not need to run this again)

```python
converted_utterances_dir = "../mmsd_raw_data/converted_utterances"
for filename in tqdm(filenames):
    old_file = f"{utterances_dir}/{filename}"
    new_file = f"{converted_utterances_dir}/{filename[:-4]}.wav"
    os.system(f"ffmpeg -i {old_file} {new_file}")
```
